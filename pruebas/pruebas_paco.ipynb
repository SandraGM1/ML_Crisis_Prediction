```python
target = pd.read_excel('TARGET.xlsx')
target
```

```python
target['Country'] = target['Country'].str.strip()
```

```python
lista_paises = target['Country'][1:56].to_list()
```

```python
df_codigo = pd.DataFrame({'country':lista_paises})
df_codigo 
```

```python
import pycountry

def get_iso3(name):
    try:
        return pycountry.countries.lookup(name).alpha_3
    except:
        return None

def get_iso_numeric(name):
    try:
        return int(pycountry.countries.lookup(name).numeric)
    except:
        return None
```

```python
df_codigo["iso3"] = df_codigo["country"].apply(get_iso3)
df_codigo["iso_numeric"] = df_codigo["country"].apply(get_iso_numeric)
```
